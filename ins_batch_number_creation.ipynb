```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('ins_2018_2020_part2_Batching.xlsx')
df.head()
```

```python
df=df[df['Batch'].notnull()]
df.shape
```

```python
all_dates = df['CreatedDate'].unique()
len(all_dates)
```

```python
all_dates[66] # total 0 - 66 indexes
```

```python
# created Batch 2 number based on the batching interval
df_batch = pd.DataFrame(columns=df.columns)
batch = 1
i=0
while i <= 63: # batch number until 65 (= 63, 64, 65)
    df_target = df[(df['CreatedDate']==all_dates[i])|
                   (df['CreatedDate']==all_dates[i+1])|
                   (df['CreatedDate']==all_dates[i+2])]
    df_target['Batch2'] = batch
    df_batch = df_batch.append(df_target, ignore_index=True)
    i += 3
    batch += 1
    
# add batch 66
df_target = df[(df['CreatedDate']==all_dates[i])]
df_target['Batch2'] = batch
df_batch = df_batch.append(df_target, ignore_index=True)
```

```python
df_batch
```

```python
target_batch = 'Batch2'
#df_batch = df[df[target_batch].notnull()]
df_batch.head()
```

```python
batches = df_batch[target_batch].unique()
for b in batches:
    df_focus = df_batch[df_batch[target_batch] == b]
    dates = df_focus['CreatedDate'].unique()
    createdDay = 0
    for i in dates:
        df_batch.loc[(df_batch['CreatedDate']==i)&(df_batch[target_batch]==b), 'CreatedDay'] = createdDay
        #df.loc[df[<some_column_name>] == <condition>, [<another_column_name>]]
        createdDay += 1
        

    
    
```
